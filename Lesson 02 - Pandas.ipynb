#### Задание 1

```python
import numpy as np
import pandas as pd
```

```python
author = pd.DataFrame({'author_id': [1, 2, 3], 
                         'author_name': ['Тургенев', 'Чехов', 'Островский']}, 
                        columns=['author_id', 'author_name'])
author
```

```python
book = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3], 
                     'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                     'price': [450, 300, 350, 500, 450, 370, 290]}, 
                    columns=['author_id', 'book_title', 'price'])
book
```

#### Задание 2

```python
authors_price = pd.merge(author, book, on='author_id', how='inner')
authors_price
```

#### Задание 3

```python
# Take 1
top5 = authors_price.sort_values(by='price', ascending=False).head(5)
top5
```

```python
# Take 2
top5 = authors_price.nlargest(5, 'price')
top5
```

#### Задание 4

```python
# Take 1
authors_groupby = authors_price.groupby('author_name')
authors_stat = pd.DataFrame({'min': authors_groupby['price'].min(),
                             'max': authors_groupby['price'].max(),
                             'mean': authors_groupby['price'].mean()})
authors_stat
```

```python
# Take 2
authors_stat = authors_price.groupby('author_name').agg({'price': ['min', 'max', 'mean']})
authors_stat
```

#### Задание 5**

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = pd.pivot_table(data=authors_price, values='price', index='author_name', columns='cover', aggfunc='sum', 
                           fill_value=0)
book_info
```

```python
book_info.to_pickle('./book_info.pkl')
book_info2 = pd.read_pickle('./book_info.pkl')
book_info.equals(book_info2)
```
